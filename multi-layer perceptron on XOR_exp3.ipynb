```python
# ---------- MLP on Iris Dataset ----------
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load Iris dataset
df = pd.read_csv("/content/cleaned Iris col removed.csv")  # Replace with your actual path

# Encode class labels: setosa=0, versicolor=1, virginica=2
le = LabelEncoder()
df['label'] = le.fit_transform(df['species'])

# Features and target
X = df[['sepal_length', 'petal_length', 'petal_width']].values
y = df['label'].values

# Feature normalization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Define MLP model
mlp_model = MLPClassifier(
    hidden_layer_sizes=(8, 4),    # Two hidden layers: 8 and 4 neurons
    activation='relu',
    solver='adam',
    learning_rate='adaptive',
    max_iter=1000,
    random_state=42
)

# Train MLP
mlp_model.fit(X_train, y_train)

# Evaluate model
y_pred = mlp_model.predict(X_test)
print("ðŸ” Iris MLP Model Results")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))

# ---------- MLP on XOR Gate ----------
import numpy as np

# Define XOR input/output
X_xor = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y_xor = np.array([0, 1, 1, 0])

# Define XOR MLP
xor_mlp = MLPClassifier(
    hidden_layer_sizes=(4,),     # One hidden layer with 4 neurons
    activation='tanh',           # Tanh works well for XOR
    solver='adam',
    learning_rate_init=0.01,
    max_iter=10000,
    random_state=1
)

# Train XOR model
xor_mlp.fit(X_xor, y_xor)

# Predict XOR output
xor_preds = xor_mlp.predict(X_xor)
print("\nðŸ” XOR Gate MLP Results")
for i in range(len(X_xor)):
    print(f"Input: {X_xor[i]} => Expected: {y_xor[i]}, Predicted: {xor_preds[i]}")
```
