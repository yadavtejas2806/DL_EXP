```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load and preprocess data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile with RMSprop
learning_rate = 0.05
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=learning_rate, rho=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train model
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy:.4f}')

# Plot Training vs Validation Loss
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Cost Function Graph')
plt.legend()

# Plot Training vs Validation Accuracy
plt.subplot(1,3,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Graph')
plt.legend()

# Plot Learning Rate vs Epoch (constant)
plt.subplot(1,3,3)
plt.plot(range(1, len(history.history['loss'])+1),
         [learning_rate]*len(history.history['loss']), label='Learning Rate')
plt.xlabel('Epochs')
plt.ylabel('Learning Rate')
plt.title('Learning Rate vs Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load and preprocess data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Initial learning rate
initial_lr = 0.001
optimizer = tf.keras.optimizers.RMSprop(learning_rate=initial_lr, rho=0.9)
model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Scheduler to decay LR
def scheduler(epoch, lr):
    return lr * 0.9  # decay by 10% each epoch

lr_callback = tf.keras.callbacks.LearningRateScheduler(scheduler)

# Train model with scheduler
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.2,
                    callbacks=[lr_callback])

# Get learning rates per epoch
lrs = [scheduler(i, initial_lr) for i in range(10)]

# Evaluate
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy:.4f}')

# Plot Loss vs Epoch
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epoch')
plt.legend()

# Plot Accuracy vs Epoch
plt.subplot(1,3,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Epoch')
plt.legend()

# Plot Learning Rate vs Epoch
plt.subplot(1,3,3)
plt.plot(range(1, 11), lrs, label='Learning Rate')
plt.xlabel('Epochs')
plt.ylabel('Learning Rate')
plt.title('Learning Rate vs Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load and preprocess MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Initial learning rate
initial_lr = 0.001
optimizer = tf.keras.optimizers.RMSprop(learning_rate=initial_lr, rho=0.9)

model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Learning rate scheduler
def scheduler(epoch, lr):
    return lr * 0.9  # decay by 10% each epoch

lr_callback = tf.keras.callbacks.LearningRateScheduler(scheduler)

# Train model
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.2,
                    callbacks=[lr_callback],
                    verbose=1)

# Final Training Accuracy
final_train_acc = history.history['accuracy'][-1]

# Evaluate on test set
loss, final_test_acc = model.evaluate(x_test, y_test, verbose=0)

# Print final results
print(f"\nFinal Training Accuracy: {final_train_acc * 100:.2f}%")
print(f"Final Testing Accuracy: {final_test_acc * 100:.2f}%")

# Get learning rates per epoch
lrs = [scheduler(i, initial_lr) for i in range(len(history.history['loss']))]

# Plot Loss vs Epoch
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epoch')
plt.legend()

# Plot Accuracy vs Epoch
plt.subplot(1,3,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Epoch')
plt.legend()

# Plot Learning Rate vs Epoch
plt.subplot(1,3,3)
plt.plot(range(1, len(lrs)+1), lrs, label='Learning Rate')
plt.xlabel('Epochs')
plt.ylabel('Learning Rate')
plt.title('Learning Rate vs Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load and preprocess MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Different learning rates to try
learning_rates = [0.001, 0.003, 0.005]

# Dictionary to store history
history_dict = {}

for lr in learning_rates:
    print(f"\nTraining with learning rate = {lr}")

    # Build model
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])

    # RMSprop optimizer with given LR
    optimizer = tf.keras.optimizers.RMSprop(learning_rate=lr, rho=0.9)

    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # Train model
    history = model.fit(x_train, y_train,
                        epochs=20,
                        batch_size=32,
                        validation_split=0.2,
                        verbose=0)

    history_dict[lr] = history

# Plot Loss vs Epoch for all LRs
plt.figure(figsize=(8,6))
for lr, history in history_dict.items():
    plt.plot(history.history['loss'], label=f"Train Loss (LR={lr})")
    plt.plot(history.history['val_loss'], linestyle='--', label=f"Val Loss (LR={lr})")

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epoch for Different Learning Rates (RMSprop)')
plt.legend()
plt.grid(True)
plt.show()
```
